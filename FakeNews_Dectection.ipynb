```python
import pandas as pd
import numpy as np
```

```python
t=pd.read_csv('True.csv')
f=pd.read_csv('Fake.csv')
```

```python
t.head()
```

```python
f.head()
```

```python
t['label']=1
f['label']=0
t.head()
```

```python
News=pd.concat([t,f],axis=0)
News.head()
```

```python
News.tail()
```

```python
News.isnull().sum()
```

```python
News=News.drop(['title','subject','date'],axis=1)
News.head()
```

```python
News=News.sample(frac=1)   #reshufling
News.head()
```

```python
News=News.sample(frac=1)
News.head()
```

```python
News.reset_index(inplace=True)
News.head()
```

```python
News.drop(['index'],axis=1,inplace=True)
News.head()
```

```python
#feature extraction
import re #regular expression model
```

```python
def word(text):
  text=text.lower()   #converting into lowercase
  text=re.sub(r'[^\w\s]','',text)
  text=re.sub(r'https?://\S+|www\.\S+','',text)  #remove urls
  text=re.sub(r'<.*?>+', '',text)  #remove html tags
  text=re.sub(r'\n','',text) #remove newline characters
  return text
```

```python
News['text']=News['text'].apply(word)
News['text']
```

```python
x=News['text']
y=News['label']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5)
```

```python
x_test.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization=TfidfVectorizer()
xv_train=vectorization.fit_transform(x_train)
xv_test=vectorization.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
#1) logistic regression model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
Lr=LogisticRegression()
```

```python
Lr.fit(xv_train,y_train)
```

```python
pred_lr=Lr.predict(xv_test)
```

```python
Lr.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
#2)Decession tree
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(xv_train,y_train)
```

```python
pred_dt=dt.predict(xv_test)
```

```python
dt_score=dt.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dt))
```

```python
#3)Random forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(xv_train,y_train)
```

```python
pred_rf=rf.predict(xv_test)
```

```python
rf_score=rf.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_rf))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier()
gb.fit(xv_train,y_train)
```

```python
pred_gb=gb.predict(xv_test)
```

```python
gb_score=gb.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_gb))
```

```python
def output_label(n):
  if n==0:
    return "This is a Fake News"
  elif n==1:
    return "This is a True News"
```

```python
def manual_testing(news):

  testing_News = {"text": [news]} # Corrected to use the input 'news' string

  new_def_test = pd.DataFrame(testing_News)

  new_def_test["text"] = new_def_test["text"].apply(word)

  new_x_test =new_def_test["text"]

  new_xv_test = vectorization.transform(new_x_test) #Assuming 'vectorization' is your vectorizer object

  pred_lr =Lr.predict(new_xv_test)

# pred_dtc = dtc.predict(new_xv_test)

  pred_gb =gb.predict(new_xv_test)

  pred_rf=rf.predict(new_xv_test)

  return "\n\nLR Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(

  output_label(pred_lr[0]), output_label(pred_gb[0]), output_label (pred_rf[0]))
```
